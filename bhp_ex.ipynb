```python
import pandas as pd
import seaborn as sn 
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("bhp.csv")
df.head()
```

```python
df.price_per_sqft.describe()
```

```python
plt.hist(df.price_per_sqft, bins=20 , rwidth = 0.8)
plt.xlabel("price_per_sqft")
plt.ylabel("count")
plt.show()
```

```python
plt.hist(df.price_per_sqft, bins=20 , rwidth = 0.8)
plt.xlabel("price_per_sqft")
plt.ylabel("count")
plt.yscale('log')
plt.show()
```

```python
#clean using percentile
percentile_99 = df.price_per_sqft.quantile(0.999)
percentile_99
```

```python
percentile_01 = df.price_per_sqft.quantile(0.001)
percentile_01
```

```python
df[(df.price_per_sqft>percentile_99)]
```

```python
df[(df.price_per_sqft<percentile_01)]
```

```python
newdf1 = df[(df.price_per_sqft<percentile_99) & (df.price_per_sqft>percentile_01)]
```

```python
newdf1
```

```python
newdf1.shape
```

```python
newdf1.shape[0] - df.shape[0]
```

```python
newdf1.head()
```

```python
newdf1.price_per_sqft.describe()
```

```python
std_de = newdf1.price_per_sqft.std()
std_de
```

```python
mean = newdf1.price_per_sqft.mean()
mean
```

```python
std_4_lower = mean - 4*std_de
std_4_lower
```

```python
std_4_higher = mean + 4*std_de
std_4_higher
```

```python
newdf2 = newdf1[(newdf1.price_per_sqft>std_4_lower) & (newdf1.price_per_sqft<std_4_higher)]
newdf2
```

```python
newdf2.shape
```

```python
sn.histplot(newdf2.price_per_sqft, kde=True)
```
